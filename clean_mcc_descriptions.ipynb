```python
joined_mcc = " ".join(filtered_mcc)
```

```python
import re
```

```python
re.findall('MCC \d{4}—*MCC Category', joined_mcc)
```

```python
'-' == '—'
```

```python
matches = re.findall(r'MCC \d{4}—.*?MCC Category', joined_mcc)
```

```python
matches
```

```python
import pandas as pd
```

```python
mcc_5399_patch = ['Small- to mid-sized merchants selling a general line of merchandise from',
 'departments that usually have separate checkout counters. Products for sale',
 'may include apparel, dry goods, hardware, housewares, home furnishings,',
 'electronics, furniture, and cosmetics. Such merchants often offer proprietary',
 'charge or credit accounts, deliver merchandise, and maintain open stocks.',
 'General merchandise stores offer merchandise similar to that sold by',
 'department stores, but operate on a much smaller scale.']
```

```python
mcc_codes = []
documents = []
unaltered_descriptions = []
for m in matches:
    mcc_code = m[0:8]
    
    # Get the short name by finding between — and TTC, but then get rid of — and TCC
    short_name = re.findall(r'—.*?TCC', m)[0][1:-4]
    
    # print(mcc_code)
    if mcc_code == 'MCC 5399':
        # import pdb; pdb.set_trace()
        description = ''.join(mcc_5399_patch)
        unaltered_description = description
        unaltered_descriptions.append(unaltered_description)
    else:
        description_raw = m.split('MCC Description ')[1]
        description = description_raw.split(' MCC Category')[0]
        unaltered_description = description
        unaltered_descriptions.append(unaltered_description)
    # Remove numbers and ()
    description = re.sub(r'[0-9().;,•]', '', description)
    # Remove MCC
    description = description.replace('MCC', '')
    mcc_codes.append({'mcc': mcc_code[-4:],
                      'short_name': short_name,
                      'description': description,
                      'unaltered_description': unaltered_description})
    documents.append(description)
```

```python
df = pd.DataFrame(mcc_codes)
```
